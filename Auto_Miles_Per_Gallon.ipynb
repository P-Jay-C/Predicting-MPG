```python
mpg_df = mpg_df.replace('?', np.nan)
```

```python
mpg_df[mpg_df.isnull().any(axis=1)]
```

```python
mpg_df.median()
```

```python
mpg_df = mpg_df.apply(lambda x: x.fillna(x.median()),axis=0)
```

```python
mpg_df['hp'] = mpg_df['hp'].astype(float)
```

```python
# sns.pairplot(mpg_df,diag_kind='kde')
```

```python
X = mpg_df.drop('mpg',1)
X
```

```python
y = mpg_df[['mpg']]
y
```

```python
from sklearn import model_selection, linear_model
from sklearn import preprocessing

X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
X_train.columns
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, model.coef_[0][idx]))
```

```python
intercpt = model.intercept_[0]

print("The intercept of our model is {}".format(intercpt))
```

```python
model.score(X_test, y_test)
```

```python
#Ordinary Least Squares

fig = plt.figure(figsize=(10,8))
sns.residplot(x=X_test['hp'], y=y_test['mpg'], color = 'green', lowess = True)

fig = plt.figure(figsize=(10,8))
sns.residplot(x=X_test['acc'], y = y_test['mpg'], color='green', lowess = True)
```

Improving our model with zscore

```python
from scipy.stats import zscore
```

```python
mpg_df_scaled = mpg_df.apply(zscore)
```

```python
mpg_df_scaled = pd.DataFrame(mpg_df_scaled, columns=mpg_df.columns)
mpg_df_scaled
```

```python
X = mpg_df_scaled.drop('mpg',1)
y = mpg_df_scaled[['mpg']]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size=0.2,random_state=1)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
accuracy = model.score(X_test, y_test)
accuracy
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient of {} is {}".format(col_name, model.coef_[0][idx]))
```

```python
intercpt = model.intercept_[0]
print("The intercept of our model is {}".format(intercpt))
```

```python
y_predict = model.predict(X_test)

plt.scatter(y_test['mpg'], y_predict)
```
